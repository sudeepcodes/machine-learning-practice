```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv('D:/workspace/MachineLearning/machine-learning-practice/Datasets/canada_per_capita_income.csv')
```

```python
df.head()
```

```python
plt.ylabel('per capita income (USD)')
plt.xlabel('year')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='x')
```

```python
reg = linear_model.LinearRegression()
```

```python
df['year']
```

```python
X = df[['year']]
X
```

```python
y = df['per capita income (US$)']
y
```

```python
reg.fit(X, y)
```
